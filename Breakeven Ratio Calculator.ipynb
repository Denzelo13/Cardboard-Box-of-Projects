```python
import random
```

```python
balance = 100
i = 0
print('Orginal: ',balance)


while i < 288:
    
    # Market Enter fee
    balance = balance * 0.99925
    #print('After entry: ',balance)
    
    
    chance = random.randint(0,100)
    
    #Lose
    if chance < 33:
        
        balance = balance * 0.99925
        print('After Loss: ',balance)
        
        i+=1
        
        
    #Win
    elif chance >= 34:
        balance = balance * 1.00135
        print('After Win: ',balance)
        i+=1
        
print('End: ', balance)
```
